```python
# yelp_params_lst = [{
#     'latitude' : '43.321600',
#     'longitude' : '11.327948',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.314554',
#     'longitude': '11.331021',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.316539',
#     'longitude': '11.337212',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.320523',
#     'longitude': '11.323225',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.328983',
#     'longitude': '11.321913',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.317187',
#     'longitude': '11.354256',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.341757',
#     'longitude': '11.305010',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.343243',
#     'longitude': '11.326733',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.336999',
#     'longitude': '11.300975',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.323781',
#     'longitude': '11.338891',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.347800',
#     'longitude': '11.326046',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.332288',
#     'longitude': '11.324176',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.341445',
#     'longitude': '11.312348',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.326764',
#     'longitude': '11.333702',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.309896',
#     'longitude': '11.315263',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.317315',
#     'longitude': '11.332998',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.296856',
#     'longitude': '11.334854',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.343407',
#     'longitude': '11.322849',
#     'radius': '1000'
# },
# {
#     'latitude' : '43.334640',
#     'longitude': '11.304803',
#     'radius': '1000'
# }]


yelp_params_lst = []
for i in range(len(df_city_bikes)):
    params = {}
    params['latitude'] = str(df_city_bikes.loc[i, 'latitude']) 
    params['longitude'] = str(df_city_bikes.loc[i, 'longitude'])
    params['radius'] = '1000'
    yelp_params_lst.append(params)

# construct headers and payload
yelp_api_key = os.environ["YELP_API_KEY"]
yelp_url = "https://api.yelp.com/v3/businesses/search"
yelp_api_key = 'Bearer %s' %yelp_api_key
headers = {
  'Authorization': yelp_api_key
}

# send request and receive response
yelp_res_json_lst = []
for params in yelp_params_lst:
    yelp_response = requests.get(yelp_url, params=params, headers=headers)
    yelp_res_json = yelp_response.json()
    yelp_res_json_lst.append(yelp_res_json)

# check the response
yelp_res_json_lst[0]
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
yelp_res_lst = []
city_index = 0
for yelp_res_json in yelp_res_json_lst:
    yelp_result = yelp_res_json['businesses']
    for i in range(len(yelp_result)):
        dict = {}
        dict['station name'] = df_city_bikes.loc[city_index, 'station name']
        dict['name'] = yelp_result[i]['name']
        dict['address'] = yelp_result[i]['location']['address1']
        dict['rating'] = yelp_result[i]['rating']
        if yelp_result[i]['categories']:
            dict['category'] = yelp_result[i]['categories'][0]['title']
        yelp_res_lst.append(dict)
    city_index += 1

yelp_res_lst
```

Put your parsed results into a DataFrame

```python
df_yelp = pd.DataFrame(yelp_res_lst)
df_yelp
```
